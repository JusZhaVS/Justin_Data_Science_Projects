```python
from datetime import datetime

d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d %I-%p')
```

```python
df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df.loc[0, 'Date'].day_name() #to do it for one entry
```

```python
df['Date'].dt.day_name() #to do it for an entire column
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
```

```python
filt = (df['Date'] >= '2020')
df[filt]
```

```python
second_filt = ((df['Date'] >= '2019') & (df['Date'] < '2020'))
df[second_filt]
```

```python
third_filt = ((df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01')))
df[third_filt]
```
